```python
from PIL import Image
import matplotlib.pyplot as plt

# Charger limage
image = Image.open("winners.jpg")

# Afficher limage
plt.imshow(image)
plt.title("Image couleur")
plt.axis("off")
plt.show() 

```

```python
# Convertir en niveaux de gris
gray_image = image.convert("L")

# Afficher
plt.imshow(gray_image, cmap='gray')
plt.title("Image en niveaux de gris")
plt.axis("off")
plt.show()
```

```python
import numpy as np

# Binarisation avec un seuil
threshold = 128
bw_image = gray_image.point(lambda x: 255 if x > threshold else 0)

plt.imshow(bw_image, cmap='gray')
plt.title("Image Noir et Blanc (Binarisation)")
plt.axis("off")
plt.show()
```

```python
# Sauvegarder limage compresse en JPEG avec qualit rduite
gray_image.save("compressed.jpg", "JPEG", quality=30)

# Recharger pour visualiser
compressed_image = Image.open("compressed.jpg")

plt.imshow(compressed_image, cmap='gray')
plt.title("Image compresse (qualit 30%)")
plt.axis("off")
plt.show()
```

```python
import cv2

# Convertir PIL -> OpenCV pour traitement couleur
image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

# Appliquer un filtre jaune (ajouter du rouge et du vert)
yellow_filter = image_cv.copy()
yellow_filter[:, :, 0] = 0  # rduire le bleu

# Afficher
plt.imshow(cv2.cvtColor(yellow_filter, cv2.COLOR_BGR2RGB))
plt.title("Filtre jaune appliqu")
plt.axis("off")
plt.show()
```
