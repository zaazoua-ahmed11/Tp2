```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

image = Image.open("winners.jpg")
image_np = np.array(image)

if image_np.ndim == 3 and image_np.shape[2] == 3:
    image_gris = 0.299 * image_np[:, :, 0] + 0.587 * image_np[:, :, 1] + 0.114 * image_np[:, :, 2]
    image_gris = image_gris.astype(np.uint8)
else:
    image_gris = image_np  


plt.imshow(image_gris, cmap='gray')
plt.title("Image  grise (avec NumPy)")
plt.axis("off")
plt.show()
```

```python
import numpy as np

seuil = 128
image_bw = (image_gris > seuil) * 255  
image_bw = image_bw.astype(np.uint8)

plt.imshow(image_bw, cmap='gray')
plt.title("Image Noir et Blanc ")
plt.axis("off")
plt.show()
```

```python
import numpy as np
image_gris.save("imageCompressee.jpg", "JPEG", quality=30)

image_compressee = Image.open("imageCompressee.jpg")
image_compressee_np = np.array(image_compressee)

plt.imshow(image_compressee_np, cmap='gray')
plt.title("Image compresse")
plt.axis("off")
plt.show()
```
