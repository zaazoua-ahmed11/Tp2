```python
from PIL import Image
import matplotlib.pyplot as plt


image = Image.open("winners.jpg")


plt.imshow(image)
plt.title("Image couleur")
plt.axis("off")
plt.show() 

```

```python

image_gris = image.convert("L")

plt.imshow(image_gris, cmap='gray')
plt.title("Image gris")
plt.axis("off")
plt.show()
```

```python
import numpy as np


seuil = 128
image_noir_blanche = image_gris.point(lambda x: 255 if x > seuil else 0)

plt.imshow(image_noir_blanche, cmap='gray')
plt.title("Image Noir et Blanc (Binarisation)")
plt.axis("off")
plt.show()
```

```python

image_gris.save("imageCompressee.jpg", "JPEG", quality=30)


imageCompressee = Image.open("imageCompressee.jpg")

plt.imshow(imageCompressee, cmap='gray')
plt.title("Image compressée ")
plt.axis("off")
plt.show()
```

```python

import cv2


image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

yellow_filter = image_cv.copy()
yellow_filter[:, :, 0] = 0  

plt.imshow(cv2.cvtColor(yellow_filter, cv2.COLOR_BGR2RGB))
plt.title("Filtre jaune appliqué")
plt.axis("off")
plt.show()
```
